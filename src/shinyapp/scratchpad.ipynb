```python
from geopy.distance import geodesic

def dist(row):
    try:
        return geodesic((row["lat"], row["lon"]), (row["prev_lat"], row["prev_lon"])).km
    except:
        return 0
```

```python
c = [c for c in geostages2.iloc[2]["geometry"].coords]
cdf = pd.DataFrame.from_records(c, columns=["prev_lon", "prev_lat", "z"])
cdf["lat"] = cdf["prev_lat"].shift(-1)
cdf["lon"] = cdf["prev_lon"].shift(-1)

cdf["_dist"] = cdf.apply(lambda x: dist(x), axis=1)

cdf["_dist2"] = cdf.apply(
    lambda x: Geodesic.WGS84.Inverse(x["prev_lat"], x["prev_lon"], x["lat"], x["lon"])[
        "s12"
    ],
    axis=1,
)
cdf["_bearing"] = cdf.apply(
    lambda x: Geodesic.WGS84.Inverse(x["prev_lat"], x["prev_lon"], x["lat"], x["lon"])[
        "azi1"
    ],
    axis=1,
)

ddz = pd.DataFrame(
    cdf.apply(
        lambda x: Geodesic.WGS84.Inverse(
            x["prev_lat"], x["prev_lon"], x["lat"], x["lon"]
        ),
        axis=1,
    )
)[0].apply(pd.Series)

ddz["cumdist"] = ddz["s12"].cumsum()


ddz[["y", "x"]] = ddz.apply(lambda x: pd.Series(divmod(x["cumdist"], 1000)), axis=1)
ddz["y"] = 20 * ddz["y"]
ddz["azid"] = ddz["azi1"].diff()
ddz["u"] = 3 * np.sin(np.deg2rad(ddz["azi1"]))
ddz["v"] = 3 * np.cos(np.deg2rad(ddz["azi1"]))

ddz["u2"] = 3 * np.sin(np.deg2rad(ddz["azid"]))
ddz["v2"] = 3 * np.cos(np.deg2rad(ddz["azid"]))

ddz2 = linestringFomCol(ddz, latcol="lat1", loncol="lon1", groupby="y")

plt.figure(figsize=(30, 30))
plt.quiver(
    ddz["x"],
    ddz["y"],
    ddz["u2"],
    ddz["v2"],
    angles="xy",
    pivot="tail",
    color="lightgrey",
    scale=100,
)
ax = plt.gca()
for i in ddz2.index:
    inset = inset_axes(
        ax,
        width=1,  # inch
        height=1,  # inch
        bbox_transform=ax.transData,  # data coordinates
        bbox_to_anchor=(-150, i - 5),  # data coordinates
        loc=3,
    )  # loc=lower left corner

    y, x = ddz2[i].xy
    inset.plot(x, y)
    inset.plot(x[0], y[0], "o")
```
