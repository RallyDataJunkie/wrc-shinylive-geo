```python
# Highlight a segment in the context of the preceding and following segments
# This is not done to a km/m scale. We would need to
# transform the latlong co-ords first to do that
# to get eg 100m either side of a 100m stretch
import matplotlib.pyplot as plt
from shapely.geometry import LineString, Point
import geopandas as gpd

# http://toblerity.org/shapely/manual.html?highlight=cut
# The cut function will find a distance along a line and then cut it at a nearby vertex
def cut(line, distance):
    """Cut a line into two separate lines at a specified distance from its starting point."""
    if distance <= 0.0 or distance >= line.length:
        return [LineString(line)]
    coords = list(line.coords)
    # Patch out the z co-ord
    coords = [(c[0], c[1]) for c in coords]
    for i, p in enumerate(coords):
        pd = line.project(Point(p))
        if pd == distance:
            return [LineString(coords[: i + 1]), LineString(coords[i:])]
        if pd > distance:
            cp = line.interpolate(distance)
            return [
                LineString(coords[:i] + [(cp.x, cp.y)]),
                LineString([(cp.x, cp.y)] + coords[i:]),
            ]

# ALSO the following accepts diatances in meters
# wrc.GeoTools.cut_line_by_distance_meters(line, 1000, 2000)

def highlightline_segment(line, start, end):
    # TO DO this needs projecting into meters
    edgedsegment = cut(
        cut(line, start - 0.01)[1],
        end - start + 0.02,
    )[0]
    segments = cut(edgedsegment, 0.01)
    firstsegment = segments[0]
    segments = [firstsegment] + cut(segments[1], segments[1].length - 0.01)
    # Segments is a list of linestring objects
    # print(segments)
    # segments=cut(line,0.1)
    gdf_segments = gpd.GeoDataFrame(list(range(len(segments))), geometry=segments)
    gdf_segments.columns = ["index", "geometry"]

    return gdf_segments


line = geostages["geometry"].iloc[2]
print(line.length)
start = 0.05
end = 0.06
print(line)
gdf_segments = highlightline_segment(line, start, end)

fig, ax = plt.subplots()
# line.plot(ax=ax, lw=6, color='gray')
ax = gdf_segments.plot(ax=ax, lw=3, color=["lightgrey", "blue", "lightgrey"])
```

```python
from geopy.distance import geodesic

def dist(row):
    try:
        return geodesic((row["lat"], row["lon"]), (row["prev_lat"], row["prev_lon"])).km
    except:
        return 0
```

```python
from shapely.geometry import Point

def XlinestringFomCol(df, latcol="lat", loncol="lon", groupby=None):
    geometry = [Point(xy) for xy in zip(df[latcol], df[loncol])]
    dfg = gpd.GeoDataFrame(df, geometry=geometry)
    if groupby is not None:
        return dfg.groupby(groupby)["geometry"].apply(lambda x: LineString(x.tolist()))
    return LineString(dfg["geometry"].tolist())
```

```python
c = [c for c in geostages2.iloc[2]["geometry"].coords]
cdf = pd.DataFrame.from_records(c, columns=["prev_lon", "prev_lat", "z"])
cdf["lat"] = cdf["prev_lat"].shift(-1)
cdf["lon"] = cdf["prev_lon"].shift(-1)

cdf["_dist"] = cdf.apply(lambda x: dist(x), axis=1)

cdf["_dist2"] = cdf.apply(
    lambda x: Geodesic.WGS84.Inverse(x["prev_lat"], x["prev_lon"], x["lat"], x["lon"])[
        "s12"
    ],
    axis=1,
)
cdf["_bearing"] = cdf.apply(
    lambda x: Geodesic.WGS84.Inverse(x["prev_lat"], x["prev_lon"], x["lat"], x["lon"])[
        "azi1"
    ],
    axis=1,
)

ddz = pd.DataFrame(
    cdf.apply(
        lambda x: Geodesic.WGS84.Inverse(
            x["prev_lat"], x["prev_lon"], x["lat"], x["lon"]
        ),
        axis=1,
    )
)[0].apply(pd.Series)

ddz["cumdist"] = ddz["s12"].cumsum()


ddz[["y", "x"]] = ddz.apply(lambda x: pd.Series(divmod(x["cumdist"], 1000)), axis=1)
ddz["y"] = 20 * ddz["y"]
ddz["azid"] = ddz["azi1"].diff()
ddz["u"] = 3 * np.sin(np.deg2rad(ddz["azi1"]))
ddz["v"] = 3 * np.cos(np.deg2rad(ddz["azi1"]))

ddz["u2"] = 3 * np.sin(np.deg2rad(ddz["azid"]))
ddz["v2"] = 3 * np.cos(np.deg2rad(ddz["azid"]))

# TO DO - use linestringFomLatLonCols in geotools.py
# ddz2 = linestringFomCol(ddz, latcol="lat1", loncol="lon1", groupby="y")
ddz2 = wrc.GeoTools.linestringFomLatLonCols(
    ddz, latcol="lat1", loncol="lon1", groupby="y"
)
plt.figure(figsize=(30, 30))
plt.quiver(
    ddz["x"],
    ddz["y"],
    ddz["u2"],
    ddz["v2"],
    angles="xy",
    pivot="tail",
    color="lightgrey",
    scale=100,
)
ax = plt.gca()
for i in ddz2.index:
    inset = inset_axes(
        ax,
        width=1,  # inch
        height=1,  # inch
        bbox_transform=ax.transData,  # data coordinates
        bbox_to_anchor=(-150, i - 5),  # data coordinates
        loc=3,
    )  # loc=lower left corner

    y, x = ddz2[i].xy
    inset.plot(x, y)
    # Add a marker at the start of the trace
    # Should probably color this green, and a red marker at end?
    inset.plot(x[0], y[0], "o", color="green")
    inset.plot(x[-1], y[-1], "o", color="red")
```

```python
from shapely.geometry import mapping

from ipyleaflet import Polyline

m = wrc.GeoTools.simple_stage_map(geostages)

def add_route_linestring_to_map(m, route, color="red", weight=4, opacity=0.8, fill=False):
    route_coords = mapping(route)["coordinates"]
    # Create a Polyline object
    # Note: ipyleaflet expects coordinates as [lat, lon] while Shapely uses [lon, lat]
    # So we need to swap the coordinates
    route_layer = Polyline(
        locations=[[c[1], c[0]] for c in route_coords],
        color=color,
        weight=weight,
        opacity=opacity,
        fill=fill,
    )

    # Add the route to the map
    m.add_layer(route_layer)

add_route_linestring_to_map(m, geostages.iloc[2]["geometry"],"green")
add_route_linestring_to_map(m, telemroute, "black")

m
```
